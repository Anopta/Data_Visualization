```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,10,0.1)
y=2*x+5
```

```python
plt.plot(x,y)
plt.show()
```

```python
#customizing the line

x=np.arange(1,10,0.1)
y=2*x+5

plt.plot(x,y, color='r')
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x=np.arange(1,10,0.1)
y1=2*x+5
y2=3*x+10
y3=4*x+12

plt.subplot(1,3,1)
plt.plot(x,y1, color='r')
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')


plt.subplot(1,3,2)
plt.plot(x,y2, color='y')
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')


plt.subplot(1,3,3)
plt.plot(x,y3, color='g')
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#bar plot

fruit={'apple': 30, 'mango': 45, 'banana': 10}
```

```python
fruit
```

```python
names=list(fruit.keys())
```

```python
names
```

```python
quantity=list(fruit.values())
```

```python
quantity
```

```python
plt.bar(names,quantity)
plt.show()
```

```python
#customizing bar plot
plt.barh(names, quantity, color='y')
plt.title('Distribution of fruits')
plt.xlabel('fruits')
plt.ylabel('Quantity')
plt.show()
```

```python
# scatter plot

x=[4,3,5,3,1,6,7,8,9,8,7,6]
a=[1,2,3,4,5,4,2,1,4,5,3,4]
b=[3,5,4,2,6,7,6,4,6,8,9,0]
```

```python
plt.figure(num=1,figsize=(10,10))
plt.scatter(x,a, s=200)
plt.scatter(x,b,s=500, marker='4')
plt.legend(['a','b'])
plt.xlabel('x')
plt.ylabel('a&b')
plt.title('X vs a&b')
plt.show()
```

```python
#histogram

data=[1,2,3,4,3,1,2,3,4,5,4,4,2,3,4,5,2,2,5,6,7,7]
plt.hist(data)
plt.show()
```

```python
#boxplot

one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,3,5,6,4,3]
three=[6,7,8,9,3,4,5,4,3]

plt.boxplot(list([one, two, three]))
plt.show()
```

```python
#boxplot

one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,3,5,6,4,3]
three=[6,7,8,9,3,4,5,4,3]

plt.violinplot(list([one, two, three]), showmedians=True, showmeans=True)
plt.grid(True)
plt.title('Distribution of data')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#pie-chart

fruit=['apple','mango','orange','grapes']
quantity=[30,45,12,100]

plt.pie(quantity, labels=fruit, autopct= '%0.1f%%', shadow=True, explode=(0.1,0.1,0.1,0.1))
plt.show()
```

```python
fruit=['apple','mango','orange','grapes']
quantity=[30,45,12,100]

pie1=plt.pie(quantity, labels=fruit, autopct= '%0.1f%%', radius=2)
pie2=plt.pie([5], colors='w', radius=1)
plt.show()
```

```python
#area plot

x=[10,20,30,40,50,60,70,80,90]
y=[1,3,2,4,5,3,2,3,4]

plt.stackplot(x,y)
plt.show()
```
